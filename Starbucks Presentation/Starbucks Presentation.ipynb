# Data Visualization

### Visualizing Caloric Distribution

Let's start by creating a histogram to visualize the distribution of calories in the Starbucks menu items.

```python
# Plotting Histogram for Caloric Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['calories'], bins=15, kde=True, color='blue')
plt.title('Caloric Distribution in Starbucks Menu Items')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.show()
```

#### Explanation:

The histogram above shows the distribution of calories in the Starbucks menu items. We can observe a notable bump just under 200 calories, suggesting that there is a group of menu items with lower calorie counts. On the other hand, the majority of the items tend to be in the 300-500 calorie range, with the highest concentration around 350 calories. This indicates that Starbucks offers a diverse range of items with varying calorie levels, but most of their menu items are in the moderate to higher calorie range.

### Comparing Nutritional Content by Menu Item Type

Next, let's use bar plots to compare the average nutritional content (fat, carb, fiber, and protein) of menu items by their type.

```python
# Grouping by Menu Item Type and Calculating Mean Nutritional Content
mean_nutrition_by_type = df.groupby('type')[['fat', 'carb', 'fiber', 'protein']].mean()

# Plotting Bar Chart
plt.figure(figsize=(10, 6))
mean_nutrition_by_type.plot(kind='bar', colormap='turbo')
plt.title('Average Nutritional Content by Menu Item Type')
plt.xlabel('Menu Item Type')
plt.ylabel('Average Nutritional Content')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

#### Explanation:

The bar chart above compares the average nutritional content of Starbucks menu items across different food categories. Here are the key insights:

- <b>Carbohydrates (Carbs):</b> Carbs are by far the most prevalent among the different food categories. We can observe that menu items in almost all categories are relatively high in carbohydrates. This suggests that carbohydrates play a significant role in the composition of Starbucks menu items.

- <b>Fiber:</b> On the other hand, fiber seems to be the lowest among the average nutritional content across food categories. This indicates that Starbucks menu items, on average, may have room for improvement in terms of fiber content.

- <b>Protein:</b> Among the menu item types, sandwiches have the highest average protein content. This suggests that sandwiches are a good choice for customers seeking higher protein options in their Starbucks meals.

These insights shed light on the nutritional distribution of Starbucks menu items based on their food categories, enabling customers to make informed choices based on their dietary preferences and needs.

### Correlation Heatmap

Let's visualize the correlation between the numerical variables (calories, fat, carb, fiber, protein) using a heatmap.

```python
# Creating a Correlation Matrix
correlation_matrix = df[['calories', 'fat', 'carb', 'fiber', 'protein']].corr()

# Plotting Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Nutritional Content')
plt.show()
```

#### Explanation:

The heatmap above shows the correlation between the numerical variables in the Starbucks menu items. Here are the key insights:

- <b>Fat:</b> Fat has the highest positive correlation with calories, with a correlation coefficient of 0.76. This suggests that menu items with higher fat content tend to be more calorically dense.

- <b>Carbohydrates (Carbs):</b> Carbs also have a positive correlation with calories, with a correlation coefficient of 0.67. This indicates that items with higher carbohydrate content are likely to have more calories.

- <b>Protein:</b> Protein has a moderate positive correlation with calories, with a correlation coefficient of 0.41. This implies that menu items with higher protein content may contribute to the caloric content, albeit to a lesser extent compared to fat and carbs.

These correlation values provide insights into the influence of different nutritional components on the caloric content of Starbucks menu items. Understanding these correlations can help customers make informed choices based on their dietary preferences and nutritional needs.

### Scatterplot: Calories vs Carbs

Lastly, let's create a scatter plot to explore the relationship between calories and protein in the menu items.

```python
# Plotting Scatter Plot: Calories vs. Carbs
plt.figure(figsize=(8, 6))
sns.scatterplot(x='calories', y='carb', data=df, color='red', alpha=0.7)
plt.title('Calories vs. Carbs in Starbucks Menu Items')
plt.xlabel('Calories')
plt.ylabel('Carbs (g)')
plt.show()
```

#### Explanation:

The scatter plot above displays the relationship between calories and carbohydrate content in Starbucks menu items. Here are the key insights:

- Calories vs. Carbs Relationship: We can observe that as the caloric content increases, there appears to be a general trend of higher carbohydrate content as well. This indicates that menu items with more calories are likely to have higher carbohydrate levels.

- Clusters in the Scatter Plot: The scatter plot reveals some clusters of data points, suggesting distinct groups of menu items based on their calorie and carbohydrate content. These clusters may represent specific categories or product lines in the Starbucks menu.

This scatter plot helps us understand the connection between calories and carbohydrates, providing valuable insights into the nutritional composition of Starbucks menu items.
